```python
import sympy as sym
from sympy.vector import CoordSys3D
from sympy.plotting import plot_parametric
```

```python
O = sym.vector.CoordSys3D(' ')
t = sym.symbols('t')
```

```python
R = t
theta = t
```

```python
r = (R*sym.cos(theta))*O.i + (R*sym.sin(theta))*O.j
```

```python
r
```

```python
plot_parametric(r.dot(O.i),r.dot(O.j))
```

```python
v = sym.diff(r,t)
```

```python
a = sym.diff(v,t)
```

```python
e_r = r/sym.sqrt(r.dot(r))
```

```python
e_theta = O.k.cross(e_r)
```

```python
sym.simplify(a.dot(e_theta))
```

```python
sym.simplify(sym.simplify(a.dot(e_r)))
```
